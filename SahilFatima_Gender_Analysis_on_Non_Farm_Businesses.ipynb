```python
df = pd.read_csv("/content/Pakistan dataset 1960 to 2020.csv")

df.head()
```

```python
#Creating new DataFrame, Delete columns containing either 50% or more than 50% NaN Values
perc = 50.0
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
df2 = df.dropna( axis=1,
                thresh=min_count)
print("Data:new data frame ready... ")
df2.shape
```

```python
import matplotlib.pyplot as plt

# Data for visualizations
education_levels = ["Literate", "Illiterate"]
education_percentages = [31.67, 68.33]

family_structure = ["Nuclear", "Joint", "Extended"]
family_percentages = [38.13, 47.92, 13.96]

economic_location = ["Indoor", "Outdoor", "Both"]
economic_location_percentages = [71.88, 14.17, 13.96]

savings_levels = ["Below 10K PKR", "11K-20K PKR", "21K-30K PKR"]
savings_percentages = [13.75, 33.96, 45.83]

work_hours = ["1-5 hours", "6-10 hours", "Above 10 hours"]
work_hours_percentages = [4.79, 43.75, 51.46]

# Creating the visualizations
fig, axs = plt.subplots(3, 2, figsize=(14, 16))
fig.suptitle("Visualizations Based on Data", fontsize=16, fontweight='bold')

# Bar Chart for Education Levels
axs[0, 0].bar(education_levels, education_percentages, color=['blue', 'orange'])
axs[0, 0].set_title("Education Levels")
axs[0, 0].set_ylabel("Percentage")
axs[0, 0].set_ylim(0, 100)

# Pie Chart for Family Structure
axs[0, 1].pie(family_percentages, labels=family_structure, autopct='%1.1f%%', colors=['gold', 'lightblue', 'pink'])
axs[0, 1].set_title("Family Structure Distribution")

# Stacked Bar Chart for Economic Activity Location
axs[1, 0].bar(economic_location, economic_location_percentages, color=['green', 'purple', 'brown'])
axs[1, 0].set_title("Economic Activity Location")
axs[1, 0].set_ylabel("Percentage")
axs[1, 0].set_ylim(0, 100)

# Bar Chart for Savings Distribution
axs[1, 1].bar(savings_levels, savings_percentages, color=['cyan', 'magenta', 'yellow'])
axs[1, 1].set_title("Savings Distribution")
axs[1, 1].set_ylabel("Percentage")
axs[1, 1].set_ylim(0, 100)

# Histogram for Work Hours Distribution
axs[2, 0].bar(work_hours, work_hours_percentages, color=['red', 'blue', 'green'])
axs[2, 0].set_title("Work Hours Distribution")
axs[2, 0].set_ylabel("Percentage")
axs[2, 0].set_ylim(0, 100)

# General layout adjustments
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
